```python
%matplotlib notebook

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

import math as m

# matplotlib plotting parameters
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.unicode'] = True

mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'Times'
mpl.rcParams['figure.titlesize'] = 'xx-large'
mpl.rcParams['axes.labelsize'] = 18
mpl.rcParams['axes.titlesize'] = 'large'
mpl.rcParams['xtick.labelsize'] = 18
mpl.rcParams['ytick.labelsize'] = 18

# подключаем пакеты для рендеринга русского текста в LaTeX
mpl.rc('text.latex', preamble=r'\usepackage{amsmath}')
mpl.rc('text.latex', preamble=r"\usepackage[utf8]{inputenc}")
mpl.rc('text.latex', preamble=r"\usepackage[russian]{babel}")
```

```python
morse0 = np.loadtxt("./danila_morse_J=0.txt")
morse12 = np.loadtxt("./danila_morse_J=12.txt")
morse24 = np.loadtxt("./danila_morse_J=24.txt")
morse36 = np.loadtxt("./danila_morse_J=36.txt")
morse48 = np.loadtxt("./danila_morse_J=48.txt")
```

```python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

```python
color = 'k'
lw = 0.5

fig = plt.figure(figsize=(10, 7.5))
ax = fig.add_subplot(111)

plt.xlim((4.0, 15.0))
plt.ylim((-210.0, 250.0))
plt.plot( morse0[:,0], morse0[:,1], color = 'k', lw = lw )
plt.plot( morse12[:,0], morse12[:,1], color = 'r', lw = lw )
plt.plot( morse24[:,0], morse24[:,1], color = 'g', lw = lw )
plt.plot( morse36[:,0], morse36[:,1], color = 'b', lw = lw )
plt.plot( morse48[:,0], morse48[:,1], color = 'y', lw = lw )

plt.ylabel('$V(r)$, см$^{-1}$')
plt.xlabel('$r$, $a_0$')

colors = ['k', 'r', 'g', 'b', 'y']
legend_elements = []
Jvalues = [0, 12, 24, 36, 48]
for value, color in zip(Jvalues, colors):
    legend_elements.append(
        Line2D([0], [0], color = color, lw = 1.0, label = 'J = {0}'.format(value))
    )

ax.legend( handles = legend_elements, prop = {'size': 12} )    

plt.show()
```

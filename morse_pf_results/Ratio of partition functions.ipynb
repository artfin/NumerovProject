```python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
```

```python
classical_pf = np.loadtxt("./classical_pf.txt" )
quantum_pf = np.loadtxt( "./quantum_pf.txt" )

temperature = classical_pf[:,0]
ratio = classical_pf[:,1] / quantum_pf[:,1]
```

```python
classical_pf2 = np.loadtxt( "./classical_pf2.txt" )
quantum_pf2 = np.loadtxt( "./quantum_pf2.txt" )

temperature = classical_pf2[:,0]
ratio = classical_pf2[:,1] / quantum_pf2[:,1]
```

```python
fig = plt.figure(figsize=(10, 7.5))
ax = fig.add_subplot(111)

plt.plot( temperature, ratio, lw = 1.0, color = "red" )
plt.hlines(xmin = temperature[0], xmax = temperature[-1], y = 1.0, color = 'blue')

plt.xlabel("Температура, K")
plt.ylabel("$Q$ / $Q_{rovib}^{q}$")

plt.xlim((temperature[0], temperature[-1]))
plt.ylim((0.9975, 1.01))
```
